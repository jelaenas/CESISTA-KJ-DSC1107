# **3. Tidy Data**
**3.1 who Dataset**
Description of Columns:
country: Country name.

iso2: Two-digit country code.

iso3: Three-digit country code.

year: Year.

Variables like new_ep_f014:

ep: TB type (e.g., rel = relapse, ep = extrapulmonary).

f: Sex (e.g., f = female).

014: Age group (e.g., 0-14 years).

Steps:

  **3.1.1 Identify the variables in the dataset.**
  
 **3.1.2 Perform a pivot operation to make the data tidy, storing the result in who2.**

**3.1.3 Separate values like new_ep_f014 into components (e.g., new, ep, f014). Remove the column containing new, and store the result in who3.**

**3.1.4 Further separate values like f014 into f and 014, storing the result in who_tidy.**

```r
# Identify the variables in the dataset
names(who)

# Perform a pivot operation to make the data tidy, storing the result in who2
who2 <- who %>%
  pivot_longer(
    cols = starts_with("new"),
    names_to = "key",
    values_to = "cases"
  )

# Separate values like new_ep_f014 into components (e.g., new, ep, f014). Remove the column containing new, and store the result in who3
who3 <- who2 %>%
  separate(key, into = c("new", "type", "sex_age"), sep = "_") %>%
  select(-new)

# Further separate values like f014 into f and 014, storing the result in who_tidy
who_tidy <- who3 %>%
  separate(sex_age, into = c("sex", "age_group"), sep = 1) %>%
  mutate(
    age_group = case_when(
      age_group == "014" ~ "0-14",
      age_group == "1524" ~ "15-24",
      age_group == "2534" ~ "25-34",
      age_group == "3544" ~ "35-44",
      age_group == "4554" ~ "45-54",
      age_group == "5564" ~ "55-64",
      age_group == "65"   ~ "65+",
      TRUE ~ age_group
    )
  )

who_tidy
```

# **3.2 Population Dataset**
**3.2.1 Identify the variables in this dataset.**

**3.2.2 Perform a pivot operation to tidy the data, storing the result in population3.**

**3.2.3 Cast the population variable to an appropriate data type, storing the result in population_tidy.**

```r
# Identify the variables in this dataset
names(population2)

# Perform a pivot operation to tidy the data, storing the result in population3
population3 <- population2 %>%
  pivot_longer(cols = `1960`:`2023`, names_to = "year", values_to = "population")

# Cast the population variable to an appropriate data type, storing the result in population_tidy
population_tidy <- population3 %>%
  mutate(
    year = as.integer(year),
    population = as.numeric(population)
  )

population_tidy
```

# **3.3 Join Datasets**
  **3.3.1 Identify the variable(s) required to join who_tidy and population_tidy.**
  
The variables needed to join who_tidy and population_tidy are country, year, and country code/iso3.

 **3.3.2 Rename columns as needed to align variable names between datasets.**

**3.3.3 Join the datasets into a tibble called tuberculosis.**


```r
# Rename columns as needed to align variable names between datasets
population_tidy_renamed <- population_tidy %>%
  rename(
  country = "Country Name",
  country_code = "Country Code")

who_tidy_renamed <- who_tidy %>%
  rename(
    country_code = "iso3"
  )

population_tidy_renamed
who_tidy_renamed

# Join the datasets into a tibble called tuberculosis
tuberculosis <- who_tidy_renamed %>%
  inner_join(population_tidy_renamed, by = c("country", "country_code", "year"))
```

# **3.4 Clean Up Data**
  **3.4.1 Remove unnecessary variables from tuberculosis.**

  **3.4.2 Filter out rows with NA values.**

  **3.4.3 Save the cleaned data back into tuberculosis.**

```r
# Remove unnecessary variables from tuberculosis
tuberculosis <- tuberculosis %>% select(-"iso2", -"Indicator Name", -"Indicator Code")
tuberculosis

# Filter out rows with NA values
# Save the cleaned data back into tuberculosis
tuberculosis <- tuberculosis %>% drop_na()
tuberculosis
```

# **4. Data Manipulation**
  4.1 Determine the total TB cases among men and women in the 21st century in the United States. Identify which sex had more cases.

```r
tb_us_21st <- tuberculosis %>%
  filter(country == "United States of America", year >= 2000) %>%
  group_by(sex) %>%
  summarise(total_cases = sum(cases, na.rm = TRUE))

male_us <- tb_us_21st %>%
  filter (sex == "m") %>%
  pull (total_cases)

female_us <- tb_us_21st %>%
  filter (sex == "f") %>%
  pull (total_cases)

tb_us_21st
if (male_us > female_us) {
  cat(sprintf("The male sex has more TB cases with %d cases to %d female cases", male_us, female_us))
} else {
  cat(sprintf("The female sex has more TB cases with %d cases to %d male cases", female_us, male_us))
}
```

**4.2 Create a new variable, cases_per_100k, representing TB cases per 100,000 people by year, sex, age group, and TB type.**

```r
tuberculosis <- tuberculosis %>%
  mutate(cases_per_100k = (cases / population) * 100000)

tuberculosis
```

**4.3 Identify:**

- The country and year with the highest cases per 100k.

- The country and year with the lowest cases per 100k.

```r
highest_cases <- tuberculosis %>%
  group_by(country, year) %>%
  summarise(total_cases_per_100k = sum(cases_per_100k, na.rm = TRUE)) %>%
  ungroup() %>%
  filter(total_cases_per_100k == max(total_cases_per_100k, na.rm = TRUE))

print("The country/counrties with the highest cases per 100k are the following:")
highest_cases

lowest_cases <- tuberculosis %>%
 group_by(country, year) %>%
  summarise(total_cases_per_100k = sum(cases_per_100k, na.rm = TRUE)) %>%
  ungroup() %>%
  filter(total_cases_per_100k == min(total_cases_per_100k, na.rm = TRUE))

print("The country/counrties with the lowest cases per 100k are the following:")
lowest_cases
```
