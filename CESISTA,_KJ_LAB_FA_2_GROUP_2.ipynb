```r
library(tidyverse)
data("starwars", package = "dplyr")
```

# **Q1 — Dataset Inspection and Missing Values (5 points)**

```r
# Display the structure of the dataset using glimpse(starwars)
glimpse(starwars)

# Identify number of observations and number of variables

n_obs <- nrow(starwars)
n_vars <- ncol(starwars)

cat(sprintf("\nNumber of observations: %d \n", n_obs))
cat(sprintf("\nNumber of variables: %d \n", n_vars))

#Compute the number of missing values in height, mass, and homeworld.
missing_height <- sum(is.na(starwars$height))
missing_mass <-sum(is.na(starwars$mass))
missing_homeworld <-sum(is.na(starwars$homeworld))

cat(sprintf("\nMissing values in height:%d\n", missing_height))
cat(sprintf("\nMissing values in mass:%d\n", missing_mass))
cat(sprintf("\nMissing values in homeworld:%d\n", missing_homeworld))
```

# **Q2 — Create a Wide Summary Table (pivot_wider) (7 points)**

```r
# Filter the dataset to include only characters with non-missing species.
# Group the data by species and gender.
# Compute the mean height for each group.
starwars_grouped <- starwars %>% drop_na(species) %>%
  group_by(species, gender) %>%
  summarise(
    mean_height = sum(height),
    .groups = 'drop'
  )

# Use pivot_wider() so that: rows represent species, columns represent gender, values are mean height

starwars_wider <- starwars_grouped %>%
  pivot_wider(
    names_from = gender,
    values_from = mean_height
  )

starwars_wider
```

# **Q3 — Convert Wide Table Back to Long Format (pivot_longer) (6 points)**

```r
# Apply pivot_longer() to convert the gender columns back into: a column named gender, a column named mean_height
starwars_longer <- starwars_wider %>%
  pivot_longer(
    cols = -species, # Columns to pivot
    names_to = "gender",         # Name of the new column for original column names
    values_to = "mean_height"        # Name of the new column for the values
  )  %>% drop_na(mean_height)

  starwars_longer
```

# **Q4 — Create New Variables and Handle Missing Data (6 points)**

```r
# From the original dataset, create a new variable
# Categorize height into: short (<170 cm), average (170–189 cm), tall (≥190 cm)
#Replace missing homeworld values with "Unknown".

starwars_new <- starwars %>%
  mutate(
    bmi = mass / ((height/100)^2),
    height_category = case_when(
      height < 170 ~ "short",
      height >= 170 & height < 190 ~ "average",
      height >= 190 ~ "tall"
    ),
    homeworld = replace_na(homeworld, "Unknown")
  ) %>%
  relocate(bmi, .after = mass) %>%
  relocate(height_category, .after = height)

#Show glimpse() of the resulting dataset.
glimpse(starwars_new)

missing_homeworld_new <- summissing_homeworld <-sum(is.na(starwars_new$homeworld))
cat(sprintf("\nMissing values in homeworld:%d\n", missing_homeworld_new))
```

# **Q5 — Unnesting and Interpretation (6 points)**

```r
# Select name and films.
# Convert the films list-column into long format.
# Count the number of film appearances per character.
# Display the top 8 characters with the most appearances.

starwars_appearances <- starwars %>%
  select(name, films) %>%
  unnest(films) %>%
  count(name, sort = TRUE) %>%
  slice_head(n = 8)

starwars_appearances

# with films not converted into long format
starwars_appearances_listed <- starwars %>%
  select(name, films) %>%
  count(name, sort = TRUE) %>%
  slice_head(n = 8)

starwars_appearances_listed
```

**In 2–3 sentences, explain why this data must be converted to long format.**

The films column is originally in list-column, and as seen in the table below if it is not converted, the films that they are in simply shows as one only. To accurately count how many films each character appears in, it is essential to convert it into long format, as it will cause for the films to appear in separate rows. So, data can then be analyzed accurately, such as counting how many film appearances each character has.
